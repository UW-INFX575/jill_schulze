# Problem Set 2: Natural Language Processing with S3 Storage

The NLTK library and the subsequent packages were used to form stop word and stemming lists. I decided to stem the text and use the stem version for all the n-gram separations, rather than stemming at the end. The collections library was implemented for a simple n-gram "counter" function. A few publically available sources were helpful in iterating through text and extracting unigrams, bigrams, and trigrams. Sources are listed as they are used in the notebook. Bigram and Trigram code snippets from "Locally Optimal" blog: http://locallyoptimal.com/blog/2013/01/20/elegant-n-gram-generation-in-python/. Another noteable option, though more complex, is from StackOverflow: http://stackoverflow.com/questions/21844546/forming-bigrams-of-words-in-list-of-sentences-python. 
In creating the final summary files for unigrams, bigrams, & trigrams, I removed punctuation and most numbers. This was a different strategy than previous counts. While in most files I have left unigrams, bigrams, and trigrams in their "original" state, I elected to clean the data--somewhat by hand--when creating the final summary files. 

```python
import csv
import numpy as np
import pandas as pd
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.tokenize import WordPunctTokenizer
from nltk.stem.porter import *
from nltk.stem.snowball import SnowballStemmer
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures
```

```python
data = open('6334220.txt', 'r')
data.read()
```

```python
#TAKE OUT STOP WORDS
stopset = set(stopwords.words('english'))

with open('6334220.txt', 'r') as text_file:
    text = text_file.read()
    tokens = nltk.word_tokenize(str(text))
    tokens = [w for w in tokens if not w in stopset]
    print tokens
```

```python
#TEST STEMMING
stemmer = PorterStemmer()
singles = []
plurals = ['comprises', 'particulates', 'sections', 'liquids']
for plural in plurals:
    singles.append(stemmer.stem(plural))
#print singles
```

```python
#STEM 1 TXT FILE for unigrams
stemmed = []
text = tokens
for word in text:
    stemmed.append(stemmer.stem(word))
print stemmed
```

```python
# MAKE DF OF STEMMED
stemmed_df = pd.DataFrame(stemmed, columns=['words'])
stemmed_df
```

```python
# COUNT OCCURRENCE OF UNIGRAMS
data = stemmed
count_uni = Counter()
for elem in data:
    count_uni[elem] += 1
countUNI = dict(count_uni)
```

```python
# MAKE DF OF UNIGRAMS
countUNI_df = pd.DataFrame.from_dict(countUNI, orient='index')
countUNI_df
#EXPORT TO CSV
countUNI_df.to_csv(path_or_buf='/Users/jillschulze/Desktop/PS2/countUNI_6334220.csv')
```

Bigram and Trigram code snippets from "Locally Optimal" blog: http://locallyoptimal.com/blog/2013/01/20/elegant-n-gram-generation-in-python/. Another noteable option, though more complex, is from StackOverflow: http://stackoverflow.com/questions/21844546/forming-bigrams-of-words-in-list-of-sentences-python.

```python
# BIGRAMS
input_list = stemmed

def find_bigrams(input_list):
  return zip(input_list, input_list[1:])

bigrams = zip(input_list, input_list[1:])
print bigrams
```

```python
# COUNT OCCURRENCE OF BIGRAMS
data = bigrams
count_bi = Counter()
for elem in data:
    count_bi[elem] += 1
dict(count_bi)
```

```python
# MAKE DF OF BIGRAMS
countBI_df = pd.DataFrame.from_dict(count_bi, orient='index')
countBI_df

#EXPORT TO CSV
countBI_df.to_csv(path_or_buf='/Users/jillschulze/Desktop/PS2/countBI_6334220.csv')
```

```python
# TRIGRAMS
input_list = stemmed

def find_ngrams(input_list, n):
  return zip(*[input_list[i:] for i in range(n)])

trigrams = zip(input_list, input_list[1:], input_list[2:])
print(trigrams)
```

```python
# COUNT OCCURRENCE OF TRIGRAMS
data = trigrams
count_tri = Counter()
for elem in data:
    count_tri[elem] += 1
dict(count_tri)
```

```python
# MAKE DF OF BIGRAMS
countTRI_df = pd.DataFrame.from_dict(count_tri, orient='index')
countTRI_df

#EXPORT TO CSV
countTRI_df.to_csv(path_or_buf='/Users/jillschulze/Desktop/PS2/countTRI_6334220.csv')
```

```python
# MAKE SUMMARY DF FOR ALL NGRAMS
df = pd.DataFrame()
summary_6334220_df = df.append([countUNI_df, countBI_df, countTRI_df])

#EXPORT TO CSV
summary_6334220_df.to_csv(path_or_buf='/Users/jillschulze/Desktop/PS2/summary_6334220.csv')
```
